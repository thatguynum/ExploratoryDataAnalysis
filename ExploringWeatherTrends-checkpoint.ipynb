```python
df_local['SMA5']=df_local.iloc[:,3].rolling(window=5).mean()
df_local['SMA10']=df_local.iloc[:,3].rolling(window=10).mean()
df_local['SMA15']=df_local.iloc[:,3].rolling(window=15).mean()
```

```python
df_global_new=df_global.iloc[46:,:].copy() #creating copy of the global data from 1797
```

```python
df_global_new['SMA5']=df_global_new.iloc[:,1].rolling(window=5).mean()
df_global_new['SMA10']=df_global_new.iloc[:,1].rolling(window=10).mean()
df_global_new['SMA15']=df_global_new.iloc[:,1].rolling(window=15).mean()
```

```python
x1=df_global_new['year']
y1=df_global_new['avg_temp']
x2=df_local['year']
y2=df_local['avg_temp']
```

<B>PLOTTING GRAPH OF AVERAGE TEMPERATURE VS TIME

```python
#plotting graphs of avg temp vs year from both global and local data
plt.figure(figsize=(20,12))
plt.xlabel('Year',fontdict={'size':16})
plt.ylabel('Temp Celsius',fontdict={'size':16})
plt.title('Global Data Average Temp vs time',fontdict={'size':18})
plt.xticks(np.arange(1797,2013,step=5),rotation=45)
plt.plot(x1,y1,color='red',linewidth=5)
plt.grid()
plt.tick_params(labelsize=15)
plt.figure(figsize=(20,12))
plt.xlabel('Year',fontdict={'size':16})
plt.ylabel('Temp Celsius',fontdict={'size':16})
plt.title('Bangalore Data Average Temp vs time',fontdict={'size':18})
plt.xticks(np.arange(1797,2013,step=5),rotation=45)
plt.plot(x2,y2,color='green',linewidth=5)
plt.tick_params(labelsize=15)
plt.grid()
```

As you can see above are two graphs plotted of Avg.temp vs year for both local data and global data. It is observed that the 
average city temperature (Bangalore) is comparitively higher than the global average temperature.

<B>PLOTTING GRAPHS OF DIFFERENT SMAs VS TIME

```python
#plotting the different SMA intervals for global and local data and identifying which SMA interval to pick up
colour=['blue','red','yellow']
plt.figure(figsize=(20,12))
for i in range(3):
    plt.grid()
    plt.title('5-10-15 years moving average temperature comparison,Global',fontdict={'size':18})
    x1=df_global_new['year']
    y1=df_global_new.iloc[:,i+2]
    plt.xlabel('Year',fontdict={'size':16})
    plt.ylabel('SMA Celsius',fontdict={'size':16})
    plt.plot(x1,y1,color=colour[i],label=(i+1)*5)
    i=i+1
    plt.tick_params(labelsize=15)
    plt.legend()
    
plt.figure(figsize=(20,12))
for i in range(3):
    plt.grid()
    plt.title('5-10-15 years moving average temperature comparison,Bangalore',fontdict={'size':18})
    x2=df_local['year']
    y2=df_local.iloc[:,i+4]
    plt.xlabel('Year',fontdict={'size':16})
    plt.ylabel('SMA Celsius',fontdict={'size':16})
    plt.plot(x2,y2,color=colour[i],label=(i+1)*5)
    plt.tick_params(labelsize=15)
    plt.legend()
    i=i+1
```

```python
#plotting a graph of SMA with interval of 15 vs Year
plt.figure(figsize=(20,12))
x1=df_global_new['year']
x2=df_local['year']
y1=df_global_new['SMA15']
y2=df_local['SMA15']
plt.grid()
plt.title('15-years moving average temperature comparison, Global Vs Bangalore',fontdict={'size':18})
plt.xlabel('Year',fontdict={'size':16})
plt.ylabel('Temp Celsius',fontdict={'size':16})
plt.plot(x1,y1,color='red',linewidth='3',label='global')
plt.plot(x2,y2,color='blue',linewidth='3',label='Bangalore')
plt.legend()
plt.tick_params(labelsize=15)
```
